# Exercise 2

Starting with your solution to exercise 2, plot three simulated time series, one for each of the cases  = 0,  = 0.8 and  = 0.98.

Use a for loop to step through the  values

```python
t = 200
alpha_values = [0.0, 0.8, 0.98]
x = np.empty(t+1)

for a in alpha_values:
    x[0] = 0
    for i in range(t):
        x[i + 1] = a * x[i] + np.random.randn()
    plt.plot(x, label = f'$\\alpha = {a}$')
    
plt.legend()
plt.show()
```
