Let’s do one more application before we turn to exercises.

In this application, we plot the balance of a bank account over time.

There are no withdraws over the time period, the last date of which is denoted by 𝑇 .

The initial balance is 𝑏0 and the interest rate is 𝑟.

The balance updates from period 𝑡 to 𝑡 + 1 according to 𝑏𝑡+1 = (1 + 𝑟)𝑏𝑡

In the code below, we generate and plot the sequence 𝑏0, 𝑏1, … ,𝑏𝑇

```python
r = 0.06 # interest rate
t = 50 # end date
b = np.empty(t+1) #an empty NumPy array, to store all b_t
b[0] = 10 # initial balance
```

```python
for i in range(t):
    b[i + 1] = (1 + r)*b[i]
    
plt.plot(b, label = 'bank balance', color = 'k')
plt.legend()
plt.show()
```

# Exercise 1

𝑥𝑡+1 = 𝛼 𝑥𝑡 + 𝜖𝑡+1 where 𝑥0 = 0 and 𝑡 = 0, … , 𝑇

The sequence of shocks {𝜖𝑡} is assumed to be IID and standard normal.

```python
t = 200
alpha = 0.9
x = np.empty(t+1)
x[0] = 0

for i in range(t):
    x[i + 1] = alpha * x[i] + np.random.randn()


plt.plot(x, label = 'time series', color = 'r')
plt.legend()
plt.show()
```

# Exercise 2

Starting with your solution to exercise 2, plot three simulated time series, one for each of the cases 𝛼 = 0, 𝛼 = 0.8 and 𝛼 = 0.98.

Use a for loop to step through the 𝛼 values

```python
t = 200
alpha_values = [0.0, 0.8, 0.98]
x = np.empty(t+1)

for a in alpha_values:
    x[0] = 0
    for i in range(t):
        x[i + 1] = a * x[i] + np.random.randn()
    plt.plot(x, label = f'$\\alpha = {a}$')
    
plt.legend()
plt.show()
```

# Exercise 3

Similar to the previous exercises, plot the time series 𝑥𝑡+1 = 𝛼 |𝑥𝑡| + 𝜖𝑡+1 where 𝑥0 = 0 and 𝑡 = 0, … , 𝑇
Use 𝑇 = 200, 𝛼 = 0.9 and {𝜖𝑡} as before.

```python
t = 200
alpha = 0.9
x = np.empty(t+1)
x[0] = 0
  
for i in range(t):
    x[i + 1] = alpha * np.abs(x[i]) + np.random.randn()
    
plt.plot(x, label = 'alpha = 0.9', color = 'g')    
plt.legend()
plt.show()
```
